```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# إرسال طلب HTTP إلى الموقع
url = "http://books.toscrape.com/"
response = requests.get(url)

# التحقق من نجاح الطلب
if response.status_code == 200:
    print("تم جلب الصفحة بنجاح!")
else:
    print(f"فشل في جلب الصفحة. الرمز: {response.status_code}")
```

```python
# إنشاء كائن BeautifulSoup لتحليل المحتوى
soup = BeautifulSoup(response.content, 'html.parser')

# العثور على جميع عناصر الكتب
books = soup.find_all('article', class_='product_pod')

# إنشاء قوائم لتخزين البيانات
titles = []
prices = []
ratings = []
```

```python
# استخراج بيانات كل كتاب
for book in books:
    # استخراج العنوان
    title = book.h3.a['title']
    titles.append(title)

    # استخراج السعر
    price = book.find('p', class_='price_color').text
    prices.append(price)

    # استخراج التقييم
    rating = book.p['class'][1]  # class الثانية تحتوي على التقييم
    ratings.append(rating)

    # طباعة البيانات لكل كتاب (اختياري)
    print(f"العنوان: {title}, السعر: {price}, التقييم: {rating}")
```

```python
# إنشاء DataFrame من البيانات المجمعة
df = pd.DataFrame({
    'Title': titles,
    'Price': prices,
    'Rating': ratings
})

# عرض البيانات
print(df.head())
```

```python
# حفظ البيانات في ملف CSV
df.to_csv('books_data.csv', index=False, encoding='utf-8')
print("تم حفظ البيانات في ملف books_data.csv")
```
